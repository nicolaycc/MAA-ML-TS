# Exploratory Data Analysis (EDA)
## Marginally Analytic Visualizations of Time Series Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/sincosjump.csv',delimiter=',')
```

```python
fig, ax = plt.subplots(figsize=(16,4))
for col in df.columns:
    ax.plot(df[col], label=col)
ax.legend()
plt.show()
```

### Histograma

```python
fig, ax = plt.subplots(figsize=(16,6))
for col in df.columns:
    ax.hist(df[col],bins=20, alpha=0.65, label=col)
ax.set_title('Histogram of the time series data')
ax.set_ylabel('Frequency')
ax.set_xlabel('Value')
ax.legend()
plt.show()
```

### Density Plot

```python
fig, ax = plt.subplots(figsize=(16,6))
for col in df.columns:
    sns.kdeplot(df[col], fill=True, alpha=0.5, ax=ax, label=col)
    
ax.set_title('Density plot of the time series data')
ax.set_ylabel('Frequency')
ax.set_xlabel('Value')
ax.legend()
plt.show()
```

## Change in Time

Para ver el cambio del tiempo podemos usar la funcin `.diff()` de pandas 

```python
df
```

```python
df_diff = df.diff()
df_diff
```

### Histograma

```python
fig, ax = plt.subplots(figsize=(16,6))
for col in df_diff.columns:
    ax.hist(df_diff[col],bins=15, alpha=0.65, label=col)
ax.set_title('Histogram of the time series data')
ax.set_ylabel('Frequency')
ax.set_xlabel('Value')
ax.legend()
plt.show()
```

### Density plot

```python
fig, ax = plt.subplots(figsize=(16,6))
for col in df_diff.columns:
    sns.kdeplot(df_diff[col], fill=True, alpha=0.5, ax=ax, label=col)
    
ax.set_title('Density plot of the time series data')
ax.set_ylabel('Frequency')
ax.set_xlabel('Value')
ax.legend()
plt.show()
```

## Lag plot

```python
lag_list=[1,5,20,50,100]

for i in lag_list:
    fig, ax = plt.subplots(figsize=(15,4))
    pd.plotting.lag_plot(df['y3'],lag=i)
    ax.set_title(f'Lag plot for lag={i}')
    plt.show()
```

```python
lag_list=[1,5,20,50,100]
color=['r','g','b']

for i in lag_list:
    fig, ax = plt.subplots(figsize=(15,4))
    for col in df.columns:        
        pd.plotting.lag_plot(df[col],lag=i,alpha=0.5,label=col)
        ax.set_title(f'Lag plot for lag={i}')
        plt.legend()
    plt.show()
```
